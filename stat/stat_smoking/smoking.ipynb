```python
import pandas as pd

df = pd.read_csv("smoking_death.csv")
df[:4]
```

```python
grouped1 = df.groupby(['smoke_cigars_pipe','smoke_both','smoke_cigarettes']).sum()
grouped1[['population','deaths']]
```

```python
grouped2 = df.groupby(['smoke_cigars_pipe','smoke_both','smoke_cigarettes']).mean()
grouped2[['population','deaths']]
```

```python
grouped1['rate'] = grouped1['deaths']/grouped1['population']
```

```python
grouped1[['population','deaths','rate']]
```

```python
grouped3 = df.groupby('age').sum()
grouped3[['population','deaths']]
```

```python
df['pct'] = df['deaths']*10. / df['population']*10.
df[:4] 
```

```python
pivot = df.pivot_table('rate',rows=['smoke_cigars_pipe','smoke_both','smoke_cigarettes'],
                       cols='age')
```
